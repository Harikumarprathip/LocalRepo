```python
# Import libraries
from pyspark.sql import SparkSession
```

```python
#start the spark session
spark=SparkSession.builder.appName('Dataframe').getOrCreate()
```

```python
spark
```

```python
# Read the dataset
# interschema=True, If false every data will be same wrong datatypes
df_pyspark=spark.read.option('header','true').csv('/content/test1.csv',inferSchema=True)
df_pyspark.show()
```

```python
#Check the data
df_pyspark.printSchema()
```

```python
# Another way to read the data
df_pyspark=spark.read.csv('/content/test1.csv',header=True,inferSchema=True)
df_pyspark.show()
```

```python
# Adding columns in dataframe
df_pyspark=df_pyspark.withColumn('Experience After 2 year',df_pyspark['Experience']+2)
df_pyspark.show()
```

```python
# Drop the columns
df_pyspark=df_pyspark.drop('Experience After 2 year')
df_pyspark.show()
```

```python
# Rename the columns
df_pyspark.withColumnRenamed('Name','New Name').show()
```
